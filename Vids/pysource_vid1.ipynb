```python
import cv2 as cv
```

```python
import numpy as np
```

```python
# define what we are capturing which here is video from webcam

cap = cv.VideoCapture(0)

#cv2.startWindowThread() 

while True:
    # takes frame from camera
    ret, frame = cap.read()
    gray_scale = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    frame2 = cv.flip(frame, 1)
    print(frame.shape)



    # shows us the frame;"frame" is title of the movie
    cv.imshow("frame", frame)
    # shows us second frame
    cv.imshow("gray_scale", gray_scale)
    cv.imshow("flipping", frame2)


    # wait 1 milisecond
    key = cv.waitKey(1)
    
    if key == 27:
        break
    
    # do again in loop --> illusion of video
    
cap.release()

#cv.waitKey(1)
# this solution enables esc key to stop us
# https://stackoverflow.com/questions/6116564/destroywindow-does-not-close-window-on-mac-using-python-and-opencv
cv.destroyAllWindows()
cv.waitKey(1)
```

```python
cap = cv.VideoCapture("red_panda_snow.mp4")

while True:
    ret, frame = cap.read()
    frame2 = cv.flip(frame, 0)
    #useful for next exercise
    print(frame.shape)
    
    gray_scale = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    #0 = vertical

    cv.imshow("frame", frame)
    cv.imshow("gray_scale", gray_scale)
    cv.imshow("flipping", frame2)

    # at 25 fps --> 30 miliseconds between frame
    key = cv.waitKey(30)
    
    if key ==27:
        break
        
cap.release()

cv.destroyAllWindows()
cv.waitKey(1)
```

```python
# define what we are capturing which here is video from webcam

cap = cv.VideoCapture(0)

#cv2.startWindowThread() 

# codec is XVID there are many codecs this is most pop
fourcc = cv.VideoWriter_fourcc(*"XVID")
# 25 is fps and 720, 1280 is wxh of fram
out = cv.VideoWriter("flipped_steve.avi", fourcc, 25, (720, 1280))

while True:
    # takes frame from camera
    ret, frame = cap.read()
    gray_scale = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    frame2 = cv.flip(frame, 1)


    # shows us the frame;"frame" is title of the movie
    cv.imshow("frame", frame)
    # shows us second frame
    cv.imshow("gray_scale", gray_scale)
    cv.imshow("flipping", frame2)

    out.write(frame2)
    # wait 1 milisecond
    key = cv.waitKey(1)
    
    if key == 27:
        break
    
    # do again in loop --> illusion of video
out.release()
cap.release()

#cv.waitKey(1)
cv.destroyAllWindows()
cv.waitKey(1)
```
